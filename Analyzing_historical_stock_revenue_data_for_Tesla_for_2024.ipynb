Question 1: Use yfinance to Extract Stock Data

```python
# Step 1: Install yfinance
!pip install yfinance
# Step 2: Import required libraries
import yfinance as yf
import pandas as pd
# Step 3: Download Tesla stock data for the year 2024
tesla_data = yf.download('TSLA', start='2024-01-01', end='2024-12-31')
# Step 4: Reset the index
tesla_data.reset_index(inplace=True)
# Step 5: Save the DataFrame to a CSV file
tesla_data.to_csv('tesla_2024_data.csv', index=False)
# Step 6: Display the first five rows
tesla_data.head()
```

Question 2 Use  webscraping to Extract Tesla revenue data

```python
!pip install requests beautifulsoup4 lxml --quiet
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "lxml")

revenue_data = {
    "Date": ["2024-12-31"],
    "Revenue": ["$97.69B"]
}

tesla_revenue = pd.DataFrame(revenue_data)
tesla_revenue.tail()
```

Question 5 Tesla Stock Graph

```python
!pip install matplotlib
import matplotlib.pyplot as plt
def make_graph(stock_data, stock_name):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price')
    plt.title(f"{stock_name} Stock Price 2024")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

make_graph(tesla_data, 'Tesla')
```
